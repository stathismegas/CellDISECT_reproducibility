```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
csv_path = '/lustre/scratch126/cellgen/team205/aa34/Arian/Dis2P/dis2p_reproducibility_clean/reproduce_benchmarks/haber/cf_results/'
split_results = {}

allout = True
if allout:
    split_type = 'allOut'
else:
    split_type = 'targetOut'

for file in os.listdir(csv_path):
    if file.endswith('delta_pearson.csv') and (split_type in file):
        split = file.split('_')[3]
        condition = file.split('_')[4]
        split_results[split] = pd.read_csv(csv_path + file, index_col=0)
        split_results[split]['Cell Type'] = split
        split_results[split]['Genes'] = split_results[split].index
        split_results[split]['Condition'] = condition
        split_results[split]['Metric'] = 'Delta Pearson'
        
df_concat_delta = pd.concat(split_results.values())

cts = list(df_concat_delta['Cell Type'].unique())
cts = sorted(cts, key=lambda x: len(x), reverse=True)

csv_path = '/lustre/scratch126/cellgen/team205/aa34/Arian/Dis2P/dis2p_reproducibility_clean/reproduce_benchmarks/haber/cf_results/'
split_results = {}
for file in os.listdir(csv_path):
    if file.endswith('pearson.csv') and (split_type in file) and ('delta' not in file):
        split = file.split('_')[3]
        condition = file.split('_')[4]
        split_results[split] = pd.read_csv(csv_path + file, index_col=0)
        split_results[split]['Cell Type'] = split
        split_results[split]['Genes'] = split_results[split].index
        split_results[split]['Condition'] = condition
        split_results[split]['Metric'] = 'Pearson'
        
df_concat_reg = pd.concat(split_results.values())

cts = list(df_concat_reg['Cell Type'].unique())
cts = sorted(cts, key=lambda x: len(x), reverse=True)

csv_path = '/lustre/scratch126/cellgen/team205/aa34/Arian/Dis2P/dis2p_reproducibility_clean/reproduce_benchmarks/haber/cf_results/'
split_results = {}
for file in os.listdir(csv_path):
    if file.endswith('emd.csv') and (split_type in file):
        split = file.split('_')[3]
        condition = file.split('_')[4]
        split_results[split] = pd.read_csv(csv_path + file, index_col=0)
        split_results[split]['Cell Type'] = split
        split_results[split]['Genes'] = split_results[split].index
        split_results[split]['Condition'] = condition
        split_results[split]['Metric'] = 'EMD'
        
df_concat_emd = pd.concat(split_results.values())

cts = list(df_concat_reg['Cell Type'].unique())
cts = sorted(cts, key=lambda x: len(x), reverse=True)
# df = df_concat.loc[['20', 'all'], ['Dis2P', 'Condition', 'Cell Type', 'Genes']]
```

```python
df_concat = pd.concat([df_concat_reg, df_concat_delta, df_concat_emd])
```

```python
df_concat
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 6), sharey=True)

GENES = '20'

df_ax = df_concat[(df_concat['Genes'] == GENES) & (df_concat['Metric'] == 'Pearson')]
sns.barplot(df_ax[['Dis2P', 'Biolord', 'scdisinfact']], ax=axes[0], errorbar='se')
for p in axes[0].patches:
    height = p.get_height()
    axes[0].text(
        p.get_x() + p.get_width() / 2 -0.2,  # x coordinate of the text
        height,  # y coordinate of the text
        f'{height:.2f}',  # text to be displayed
        ha='center',  # horizontal alignment
        va='bottom'  # vertical alignment
    )
axes[0].set_title('Pearson Corr.')

df_ax = df_concat[(df_concat['Genes'] == GENES) & (df_concat['Metric'] == 'Delta Pearson')]
sns.barplot(df_ax[['Dis2P', 'Biolord', 'scdisinfact']], ax=axes[1], errorbar='se')
for p in axes[1].patches:
    height = p.get_height()
    axes[1].text(
        p.get_x() + p.get_width() / 2 -0.2,  # x coordinate of the text
        height,  # y coordinate of the text
        f'{height:.2f}',  # text to be displayed
        ha='center',  # horizontal alignment
        va='bottom'  # vertical alignment
    )
axes[1].set_title('Delta Pearson Corr.')

df_ax = df_concat[(df_concat['Genes'] == GENES) & (df_concat['Metric'] == 'EMD')]
sns.barplot(df_ax[['Dis2P', 'Biolord', 'scdisinfact']], ax=axes[2], errorbar='se')
for p in axes[2].patches:
    height = p.get_height()
    axes[2].text(
        p.get_x() + p.get_width() / 2 -0.2,  # x coordinate of the text
        height,  # y coordinate of the text
        f'{height:.2f}',  # text to be displayed
        ha='center',  # horizontal alignment
        va='bottom'  # vertical alignment
    )
axes[2].set_title('Earth Moving Distance')

# Adding labels and legend
axes[0].set_ylabel('Value')
# axes[1].legend(loc='upper left', bbox_to_anchor=(1, 1))
axes[0].spines[['right', 'top']].set_visible(False)
axes[1].spines[['right', 'top']].set_visible(False)
axes[2].spines[['right', 'top']].set_visible(False)

fig.suptitle('Haber Counterfactuals (Control --> Salmonella/Hpoly10) | Mean over splits (Cell Types) | Top 20 DEGs', fontsize=16)
plt.tight_layout()
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0, hspace=None)

plt.show()
```
