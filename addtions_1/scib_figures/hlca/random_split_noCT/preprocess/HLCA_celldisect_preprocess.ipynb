```python
def create_age_category(age):
    if age < 10:
        return '0-9'
    elif age < 20:
        return '10-19'
    elif age < 30:
        return '20-29'
    elif age < 40:
        return '30-39'
    elif age < 50:
        return '40-49'
    elif age < 60:
        return '50-59'
    elif age < 70:
        return '60-69'
    elif age < 80:
        return '70-79'
    else:
        return '80+'
```

```python
age_counts = adata.obs["age_or_mean_of_age_range"].value_counts()

valid_ages = age_counts[age_counts >= 200].index

adata = adata[adata.obs["age_or_mean_of_age_range"].isin(valid_ages)]
```

```python
adata.obs['age_category'] = adata.obs["age_or_mean_of_age_range"].apply(create_age_category)

adata.obs['age_category'] = adata.obs['age_category'].astype('category')
```
